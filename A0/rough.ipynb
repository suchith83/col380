```python
# take cache references and cache misses and calculate the miss rate to get the hit rate

df['hit_rate'] = 1 - (df['Cache Misses'] / df['Cache References'])
```

```python
# take average of hit rate over type,

df.groupby('Type')['hit_rate'].mean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Read the CSV file
data = pd.read_csv('data_1.csv')

# Extract unique loop types and sizes
loop_types = data['loop_type'].unique()
matrix_sizes = data['size'].unique()

# Plot the data
plt.figure(figsize=(10, 6))

for loop in loop_types:
    loop_data = data[data['loop_type'] == loop]
    plt.plot(loop_data['size'], loop_data['time'], marker='o', label=f'Loop {loop}')

# Add labels, title, legend, and grid
plt.xlabel('Matrix Size', fontsize=12)
plt.ylabel('Execution Time (seconds)', fontsize=12)
plt.title('Execution Time vs Matrix Size for Different Loop Permutations', fontsize=14)
plt.legend(title="Loop Type", fontsize=10)
plt.grid(alpha=0.6)

# Save and show the plot
plt.savefig('section_1.png', dpi=300)
plt.show()
```

```python

# Read the new CSV file
data2 = pd.read_csv('data2.csv')

# Extract unique permutations and matrix sizes
permutations = data2['permutation'].unique()
matrix_sizes = data2['size'].unique()

# Plot the data
plt.figure(figsize=(10, 6))

for perm in permutations:
    perm_data = data2[data2['permutation'] == perm]
    plt.plot(perm_data['size'], perm_data['user_time'], marker='o', label=f'Permutation {perm}')

# Add labels, title, legend, and grid
plt.xlabel('Matrix Size', fontsize=12)
plt.ylabel('User Time (seconds)', fontsize=12)
plt.title('Execution Time vs Matrix Size for Different Permutations (Section 2)', fontsize=14)
plt.legend(title="Permutation", fontsize=10)
plt.grid(alpha=0.6)

# Save and show the plot
plt.savefig('section_2.png', dpi=300)
plt.show()

# Analysis: Compare Section 1 and Section 2
# Load section 1 data for comparison (if available)
# data1 = pd.read_csv('data_1.csv')

# # Compare and analyze differences in user times
# for size in matrix_sizes:
#     times1 = data1[data1['size'] == size]['time']
#     times2 = data2[data2['size'] == size]['user_time']
#     # print(f"Matrix Size: {size}")
#     # print(f"Section 1 Times: {list(times1)}")
#     # print(f"Section 2 Times: {list(times2)}")
#     # print("Difference in Times:", list(times2 - times1))
#     # print("-" * 50)
```
